#### –ó–∞–¥–∞–Ω–∏–µ 1
–°–ª—É—á–∞–π–Ω–∞—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ A –∏–º–µ–µ—Ç —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ (200, 800].
–ù–∞–π–¥–∏—Ç–µ –µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é.

```python
def msm(a, b):
    mx = (a + b)/2
    dx = (b - a)**2/12
    print(f'–°—Ä–µ–¥–Ω–µ: {mx}, –î–∏—Å–ø–µ—Ä—Å–∏—è: {dx}')

res = msm(200, 800)
```

#### –ó–∞–¥–∞–Ω–∏–µ 2

–û —Å–ª—É—á–∞–π–Ω–æ–π –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω–µ B –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –µ–µ –¥–∏—Å–ø–µ—Ä—Å–∏—è —Ä–∞–≤–Ω–∞ 0.2. –ú–æ–∂–Ω–æ –ª–∏ –Ω–∞–π—Ç–∏ –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –≤–µ–ª–∏—á–∏–Ω—ã B –∏ –µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–Ω–∞—è, —á—Ç–æ –ª–µ–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ —Ä–∞–≤–Ω–∞ 0.5? –ï—Å–ª–∏ –¥–∞, –Ω–∞–π–¥–∏—Ç–µ –µ–µ.

```python
from math import sqrt
def msm(a, b):
    s2 = sqrt(a * 12) + b
    mx2 = (b + s2) / 2
    print(f'–ü—Ä–∞–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞: {round(s2, 3)}, –ï—ë —Å—Ä–µ–¥–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞: {round(mx2, 3)}')

res = msm(0.2, 0.5)
```
