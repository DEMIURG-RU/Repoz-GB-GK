```python
plt.hist(df[df['Class'] == 1]['V1'], density = True, bins = 20, alpha = 0.5, color = 'red')
plt.hist(df[df['Class'] == 0]['V1'], density = True, bins = 20, alpha = 0.5, color = 'grey')
plt.legend(labels = ['Class 1', 'Class 0'])
plt.xlabel('V1')
plt.show
```
